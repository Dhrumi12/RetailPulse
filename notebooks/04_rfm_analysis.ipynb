```python
import pandas as pd

df = pd.read_csv("../data/processed/cleaned_retail.csv")

df.head()
```

```python
df["Revenue"] = df["Quantity"] * df["UnitPrice"]

df.head()
```

```python
df["InvoiceDate"] = pd.to_datetime(
    df["InvoiceDate"]
)
```

```python
snapshot_date = (
    df["InvoiceDate"].max()
    + pd.Timedelta(days=1)
)

print(snapshot_date)
```

```python
rfm = df.groupby(
    "CustomerID"
).agg({
    "InvoiceDate":
    lambda x:
    (snapshot_date - x.max()).days,

    "InvoiceNo":
    "nunique",

    "Revenue":
    "sum"
})
```

```python
rfm.columns = [
    "Recency",
    "Frequency",
    "Monetary"
]

rfm.head()
```

```python

rfm.describe()
```

```python
rfm.to_csv("../data/processed/rfm_customers.csv")
```
