```python
import pandas as pd
```

```python
sales = pd.read_csv(
    "../data/processed/daily_sales.csv"
)

sales.head()
```

```python
forecast = pd.read_csv(
    "../data/processed/forecast_30_days.csv"
)

forecast.head()
```

```python
customers = pd.read_csv(
    "../data/processed/customer_segments.csv"
)

customers.head()
```

```python

print("Sales:", sales.shape)
print("Forecast:", forecast.shape)
print("Customers:", customers.shape)
```

```python
total_revenue = sales["Revenue"].sum()

avg_daily_revenue = sales["Revenue"].mean()

print("Total Revenue:", total_revenue)
print("Average Daily Revenue:", avg_daily_revenue)
```

```python

sales.to_csv(
    "../dashboard/sales_dashboard.csv",
    index=False
)

forecast.to_csv(
    "../dashboard/forecast_dashboard.csv",
    index=False
)

customers.to_csv(
    "../dashboard/customer_dashboard.csv",
    index=False
)
```
